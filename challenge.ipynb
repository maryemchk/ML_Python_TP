```python
# Importation des bibliothèques nécessaires
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from scipy.spatial import distance

# 1 - Charger le dataset iris dans un DataFrame nommé iris_challenge
iris_data = load_iris()
iris_challenge = pd.DataFrame(data=iris_data['data'], columns=iris_data['feature_names'])
iris_challenge['species'] = iris_data['target']

# 2 - Afficher la première ligne de iris_challenge
print("Première ligne du dataset iris_challenge:")
print(iris_challenge.iloc[0])

# 3 - Calculer la distance Euclidienne entre une plante avec longueur de sépale de 5 cm et la première plante du dataset iris
reference_plant = [5, iris_challenge.iloc[0, 1], iris_challenge.iloc[0, 2], iris_challenge.iloc[0, 3]]
first_plant = iris_challenge.iloc[0, :-1].values
euclidean_distance_first = distance.euclidean(reference_plant, first_plant)

print(f"Distance Euclidienne entre la plante de référence et la première plante du dataset: {euclidean_distance_first}")

# 4 - Calculer la distance Euclidienne entre chaque plante du dataset et une plante de référence (longueur de sépale = 5, largeur de sépale = 3.5, longueur de pétale = 1.4, largeur de pétale = 0.2)
ref_plant = [5, 3.5, 1.4, 0.2]

# Utiliser la méthode apply pour calculer la distance Euclidienne pour chaque ligne du dataset
distances = iris_challenge.apply(lambda row: distance.euclidean(ref_plant, row[:-1]), axis=1)

# Retourner un nouvel objet Series contenant les valeurs de distance
print("Distances calculées entre chaque plante et la plante de référence:")
print(distances)

# 5 - Compléter la fonction predict_petal_length qui utilise la technique des k plus proches voisins (KNN)
def predict_petal_length(new_plant):
    # Copier le DataFrame iris_challenge
    temp_df = iris_challenge.copy()

    # Calculer la distance Euclidienne entre chaque plante du dataset et new_plant
    temp_df['distance'] = temp_df.apply(lambda row: distance.euclidean(new_plant, row[:-1]), axis=1)

    # Trier sur la colonne distance et sélectionner les 5 premières plantes
    nearest_neighbors = temp_df.nsmallest(5, 'distance')

    # Calculer la moyenne des longueurs de pétale des 5 plus proches voisins
    mean_petal_length = nearest_neighbors['petal length (cm)'].mean()

    return mean_petal_length

# 5 - Utiliser la fonction predict_petal_length pour prédire la longueur de pétale d'une plante
new_plant_characteristics = [5, 3.5, 0.2]  # Caractéristiques de la nouvelle plante
predicted_petal_length = predict_petal_length(new_plant_characteristics)

print(f"La longueur de pétale prédite pour la nouvelle plante est: {predicted_petal_length} cm")
```
