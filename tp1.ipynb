```python
# désactive les avertissements liés aux assignations chaînées dans le DataFrame,
#ce qui permet de modifier des sous-ensembles de données sans recevoir d'alertes potentiellement utiles:
pd.options.mode.chained_assignment = None
np.random.seed(1)

train_df = paris_listings.iloc[0:6000]
test_df = paris_listings.iloc[6000:]

def predict_price(new_listing):
    temp_df=train_df.copy()
    temp_df['distance']=temp_df['accommodates'].apply(lambda x:np.abs(x-new_listing))
    temp_df=temp_df.sort_values('distance')
    nearest_neighbor_prices=temp_df.iloc[0:5]['price']
    predicted_price=nearest_neighbor_prices.mean()
    return(predicted_price)
        

    
```

```python
test_df['predicted_price']=test_df['accommodates'].apply(lambda x: predict_price(x))
```

```python
from sklearn.metrics import mean_squared_error
mae = np.mean(np.absolute(test_df['predicted_price'] - test_df['price']))
print("MAE (Accommodates):", mae)

# Calculer MSE
mse = np.mean((test_df['predicted_price'] - test_df['price']) ** 2)
print("MSE (Accommodates):", mse)
rmse=np.sqrt(mse)

```

```python
def predict_price_bedrooms(new_listing):
    temp_df = train_df[train_df['bedrooms'] == new_listing]
    predicted_price = temp_df['price'].median() if not temp_df.empty else train_df['price'].median()
    return predicted_price

mse = mean_squared_error(test_df['price'], test_df['predicted_price'])
print("mse:",mse)

test_df['predicted_price'] = test_df['bedrooms'].apply(lambda x: predict_price_bedrooms(x))

test_df['squared_error'] = (test_df['predicted_price'] - test_df['price']) ** 2

mse = np.mean(test_df['squared_error'])
print("MSE (Bedrooms):", mse)

rmse = np.sqrt(mse)
print("RMSE (Bedrooms):", rmse)
```

```python
# Définir les séries d'erreurs
errors_one = pd.Series([5, 10, 5, 10, 5, 10, 5, 10, 5, 10, 5, 10, 5, 10, 5, 10, 5, 10])
errors_two = pd.Series([5, 10, 5, 10, 5, 10, 5, 10, 5, 10, 5, 10, 5, 10, 5, 10, 5, 1000])

# Calculer MAE pour errors_one
mae_one = np.mean(np.abs(errors_one))
print("MAE for errors_one:", mae_one)

# Calculer RMSE pour errors_one
rmse_one = np.sqrt(np.mean(errors_one ** 2))
print("RMSE for errors_one:", rmse_one)

# Calculer MAE pour errors_two
mae_two = np.mean(np.abs(errors_two))
print("MAE for errors_two:", mae_two)

# Calculer RMSE pour errors_two
rmse_two = np.sqrt(np.mean(errors_two ** 2))
print("RMSE for errors_two:", rmse_two)
```
